Data Preprocessing for a Retail Dataset
Objective: To understand the process of data preprocessing including data integration, data cleaning,
and data transformation using a retail dataset. This case study will help students learn the practical
aspects of preparing data for analysis, ensuring data quality, and deriving meaningful insights.
Dataset: A sample retail dataset containing information about transactions made by customers at a
chain of stores. The dataset includes the following columns:
1. TransactionID: Unique identifier for each transaction
2. CustomerID: Unique identifier for each customer
3. ProductID: Unique identifier for each product
4. Quantity: Quantity of the product purchased
5. Price: Price of the product
6. Order Date: Date of the transaction
7. CustomerLocation: location of the customer

```python
import pandas as pd
```

```python
transac = pd.read_excel("ecommerce_dataset.xlsx", sheet_name='ecommerce_dataset')
prod = pd.read_excel("ecommerce_dataset.xlsx",  sheet_name='ecommerce_dataset(1)')
```

Perform the integration of the Transactions and Products tables and display the first 5
rows of the integrated dataset.

```python
integrated_data = pd.merge(transactions, products, on='ProductID')
print("Integrated Dataset:")
print(integrated_data.head())
```

```python
integrated_data
```

Write a Python code snippet to identify and remove duplicate records from the dataset.

```python
integrated_data = integrated_data.drop_duplicates()
```

```python
integrated_data
```

How would you handle outliers in the Price column using the interquartile range (IQR)
method?

```python
Q1 = integrated_data['OrderPrice'].quantile(0.25)
Q3 = integrated_data['OrderPrice'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
integrated_data = integrated_data[(integrated_data['OrderPrice'] >= lower_bound) & (integrated_data['OrderPrice'] <= upper_bound)]
```

```python
integrated_data
```

Perform Min-Max normalization on the Quantity and Price columns and display the
transformed dataset.

```python
integrated_data['OrderQuantity'] = (integrated_data['OrderQuantity'] - integrated_data['OrderQuantity'].min()) / (integrated_data['OrderQuantity'].max() - integrated_data['OrderQuantity'].min())
integrated_data['OrderPrice'] = (integrated_data['OrderPrice'] - integrated_data['OrderPrice'].min()) / (integrated_data['OrderPrice'].max() - integrated_data['OrderPrice'].min())
```

```python
integrated_data
```

Apply one-hot encoding to the Category column and show the updated dataset.

```python
integrated_data = pd.get_dummies(integrated_data, columns=['Category'])
print("Transformed Dataset:")
print(integrated_data.head())
```
